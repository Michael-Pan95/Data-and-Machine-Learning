```python
#inspect what value a data set could be for gender
import csv
f = open('legislators.csv')
legislators = list(csv.reader(f))
legislators = legislators[1:-1]
gender = []
for g in legislators:
    gender.append(g[3])
gender = set(gender)
print(gender)
```

### Missing value is common. The solution could be:
- Remove any rows that contain missing data.
- Populate the empty fields with a specified value.
- Populate the empty fields with a calculated value.
- Use analysis techniques that work with missing data.

```python
# Populate the empty fields with a specified value.
f = open('legislators.csv')
gender = list(csv.reader(f))
for g in legislators:
    if g[3] == '':
        g[3] = 'M'
print(gender)
```

## Use Error handle to deal with missing value

```python
# add new column that represents the year of birth
for b in legislators:
    try:
        birth_year = int(b[2].split('-')[0])
    except Exception:
        birth_year = 0
    b.append(birth_year)
print(legislators)
```

## Analysis techniques that work with missing data
Notice that the birth year is almost ordered in chronological order. we can replace the missing value with the birth year in last row

```python
last_value = 1
for y in legislators:
    if y[7] == 0:
        y[7] = last_value
    last_value = y[7]
print(legislators)
```
